```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#strip plot
var= sns.load_dataset("tips")
```

```python
var
```

```python
sns.stripplot(x="day" , y="total_bill",data=var, hue="sex")
```

```python
sns.stripplot(x="day" , y="total_bill",data=var, hue="sex", jitter=3)
```

```python
sns.stripplot(x="day" , y="total_bill",data=var, hue="sex", jitter=3, size=4)
```

```python
sns.stripplot(x=var["total_bill"])
```

```python
sns.stripplot(y=var["total_bill"])
```

```python
#box plot 
var= sns.load_dataset("tips")
```

```python
var
```

```python
sns.set(style="whitegrid")
```

```python
sns.boxplot(x="day", y="total_bill", data=var, hue="time")
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=var, hue="time", showmeans= True)
plt.show()
```

```python
sns.boxplot(x="day", y="total_bill", data=var, hue="time", showmeans= True, linewidth=3, palette="plasma")
plt.show()
```

```python
sns.boxplot(x="size", y="total_bill", data=var, hue="time", showmeans= True, linewidth=3, palette="plasma", orient="h")#both numeric parameter
plt.show()
```

```python
#factor plot 
sns.catplot(x="size", y="tip", data=var, hue="sex", )
plt.show()
```

```python
sns.catplot(x="size", y="tip", data=var, hue="sex",kind="box")
plt.show()
```
