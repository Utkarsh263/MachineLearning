# Sklearn Pipelines

Pipeline with chains together multiple steps so that the output of each step is used as input to the next step.
It make it easy to apply the same processing to train and test

### With Pipelines

```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.pipeline import Pipeline, make_pipeline
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.tree import DecisionTreeClassifier
```

```python
df= pd.read_csv("C:\\Users\\utkar\\Downloads\\train.csv")
```

```python
df.drop(columns=['PassengerId', 'Name','Ticket','Cabin'], inplace=True)
```

```python
df.sample(10)
```

```python
X_train,X_test,y_train,y_test= train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.2, random_state=42)
```

```python
#imputation transformer
trf1= ColumnTransformer([
    ('impute_age',SimpleImputer(),[2]),
    ('impute_embarked',SimpleImputer(strategy='most_frequent'),[6])
], remainder='passthrough')
```

```python
#onehot encoding
trf2= ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])
],remainder='passthrough')
```

```python
#Scaling 
trf3= ColumnTransformer(['scale', MinMaxScaler(), slice(0,10)])
```

```python
#Feature Scaling 
trf4= SelectKBest(score_func= chi2, k=8)
```

```python
trf5= DecisionTreeClassifier()
```

### Create Pipeline

```python
pipe= Pipeline([
    ('trf1',trf1),
    ('trf2',trf2),
    ('trf3',trf3),
    ('trf4',trf4),
    ('trf5',trf5)
])
```
