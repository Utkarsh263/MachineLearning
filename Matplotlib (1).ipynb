```python
import matplotlib.pyplot as plt
x=['Mon', 'Tues','Wed', 'Thu','Fri','Sat','Sun']
y=[12,3,11,10,7,15,2]
plt.plot(x,y)
plt.title("Sales of Stocks in a week ")
plt.xlabel("Days of the week")
plt.ylabel("Number of Stocks")
plt.show()
```

```python
#plt.plot(x , y , color="colorname" , linestyle = value , linewidth= number, marker = "markersymbol" , label= value)
months= [1,2,3,4]
sales=[12,10,18,15]
plt.plot(months, sales, color='blue',linestyle= "--", linewidth =2, marker="o", label="Stocks sales in a month")
plt.title("Sales of Stocks")
plt.xlabel("Months")
plt.ylabel("Sales per month of Stocks")
plt.legend() #fontsize , loc= upper-left/lower-right(optional)
plt.grid(color="red" , linestyle=":", linewidth=1)
plt.xlim(1,4)
plt.ylim(0,20)
plt.xticks([1,2,3,4],['M1','M2','M3','M4']) #Optional
plt.show()
```

```python
#Barcharts for category comparison and analysis
#plt.bar(x , height, color="color_name" , width= value , label= "labelname")
x=['A','B','C','D']
y=[100,250,75,128]
plt.bar(x,y,label="Sales-2025") #for horizontal use hbar
plt.title("Sales anlaysis")
plt.xlabel("Product")
plt.ylabel("Number of sales")
plt.legend()
#plt.grid()
plt.show()
```

```python
#Pie Chart uses for Proportional Illustratiion
#plt.pie(values , labels= label_list , colors=color_list, autopct='%1.1f%%')
regions=['North', 'South', 'East','West']
revenue=[10000,12000,4000,6000]
plt.pie(revenue, labels = regions,colors=['gold','teal','coral','lightgreen'],autopct='%1.1f%%')
plt.title("Contribution by each region")
plt.show()
        
```

```python
#histograms for frequency distribution represented by bins known as frequency 
#plt.hist(data, bins= number of bins, color="colorname", edgecolor="black"
marks=[12,23,45,67,89,12,34,56,76,43,67,89,10,21,34,56,72,87,11,56,89,99,100,11,22]
plt.hist(marks, bins=8, color="gold", edgecolor="black", label="Number of Students")
plt.xlabel("Marks Range")
plt.ylabel("Number of students")
plt.title("Marks Distribution of a class")
plt.legend()
#plt.grid()
plt.show()
```

```python
#Scatterplots find corelation between 
#plt.scatter(x,y,color=color_nmae, style=style-name,label=label_name)
hours_studied=[1,2,3,4,5,6,7,8,9]
marks_obtained =[30,35,40,45,50,55,60,65,70]
plt.scatter(hours_studied, marks_obtained, marker='o',color='red',label="Student's data")
plt.xlabel("Number of hours studied")
plt.ylabel("Marks obtained")
plt.title("Trend between Hours studied and Marks obtained")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.scatter([1,2,3],[35,50,60], color='blue', marker='o',label='Class A')
plt.scatter([1,2,3],[45,70,80], color='orange', marker='o',label='Class B')
plt.xlabel("Number of hours studied")
plt.ylabel("Marks obtained")
plt.title("Comaprison between Class A and Class B")
plt.legend()
plt.grid()
plt.show()
```

```python
#Subplots Function 
#plt.subplot(nrows, ncols, index(1 based indexing)) 1st method 
x=[1,2,3,4]
y=[12,20,15,25]

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("Line Chart")

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title("Bar Chart")

plt.tight_layout()
plt.show()
```

```python
#fig , ax =plt.subplots(nrows , ncols, figsize=(width , height))
fig , ax =plt.subplots(1,2, figsize=(10,5))
x=[1,2,3,4]
y=[10,20,15,28]
ax[0].plot(x,y,color='red')
ax[0].set_title("Line Chart")

ax[1].bar(x,y,color='green')
ax[1].set_title("Bar Chart")

fig.suptitle("Comparison between Line and Bar Chart")

plt.tight_layout()#always above show command
plt.show()

```
