```python
#barplot
var= sns.load_dataset("penguins")
```

```python
print(var)
```

```python
sns.barplot(x= "island" , y="bill_length_mm", data=var, hue="sex")
plt.show()
```

```python
order_1= ['Dream', 'Torgersen','Biscoe']
sns.barplot(x= "island" , y="bill_length_mm", data=var, hue="sex", order= order_1)
plt.show()
```

```python
order_1= ['Dream', 'Torgersen','Biscoe']
sns.barplot(x= "island" , y="bill_length_mm", data=var, hue="sex", order= order_1, hue_order=['Female','Male'])
plt.show()
```

```python
order_1= ['Dream', 'Torgersen','Biscoe']
sns.barplot(x= "island" , y="bill_length_mm", data=var, hue="sex", order= order_1, hue_order=['Female','Male'], ci=90)
plt.show()
```

```python
sns.barplot(x= "bill_depth_mm" , y="bill_length_mm", data=var,orient = 'h')
plt.show()
```

```python
order_1= ['Dream', 'Torgersen','Biscoe']
sns.barplot(x= "island" , y="bill_length_mm", data=var, hue="sex", order= order_1, hue_order=['Female','Male'],saturation=10,errcolor='green',errwidth=1, capsize=0.3,dodge=True,alpha=0.7,ci=90)
plt.show()
```

```python
#scatter plot 
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var, hue="sex", style="sex", size="sex", sizes=(100,40), palette="Oranges", alpha=0.75)
plt.show()
```

```python
m={"Male":"*","Female":"o"}
sns.scatterplot(x="bill_length_mm",y="bill_depth_mm",data=var, hue="sex", style="sex", size="sex", sizes=(100,40), palette="Oranges", alpha=0.75, markers=m)
plt.show()
```
